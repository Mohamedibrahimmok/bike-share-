```python
import pandas as pd
```

```python
df = pd.read_excel('1.-Badly-Structured-Sales-Data-1.xlsx')
df
```

```python
df = df.fillna(0)
df
```

```python
df.rename(columns = {'Segment>>':'Order ID','Unnamed: 2':'Consumer Same Day', 'Consumer':'Consumer First Class', 'Unnamed: 3':'Consumer Second Class', 'Unnamed: 4':'Consumer Standard Class', 'Corporate':'Corporate First Class', 'Unnamed: 7':'Corporate Same Day', 'Unnamed: 8':'Corporate Second Class', 'Unnamed: 9':'Corporate Standard Class', 'Home Office':'Office First Class', 'Unnamed: 12':'Office Same Day', 'Unnamed: 13':'Office Second Class', 'Unnamed: 14':'Office Standard Class'}, inplace = True)
```

```python
df.drop(df.index[0:2], inplace=True)
```

```python
df['Consumer First Class']=df['Consumer First Class'].astype(float, errors = 'raise')
df['Consumer Same Day']=df['Consumer Same Day'].astype(float, errors = 'raise')
df['Consumer Standard Class']=df['Consumer Standard Class'].astype(float, errors = 'raise')
df['Corporate First Class']=df['Corporate First Class'].astype(float, errors = 'raise')
df['Corporate Same Day']=df['Corporate Same Day'].astype(float, errors = 'raise')
df['Corporate Second Class']=df['Corporate Second Class'].astype(float, errors = 'raise')
df['Corporate Standard Class']=df['Corporate Standard Class'].astype(float, errors = 'raise')
df['Office First Class']=df['Office First Class'].astype(float, errors = 'raise')
df['Office Same Day']=df['Office Same Day'].astype(float, errors = 'raise')
df['Office Second Class']=df['Office Second Class'].astype(float, errors = 'raise')
df['Office Standard Class']=df['Office Standard Class'].astype(float, errors = 'raise')
```

```python
df['Segment']=0
df['Ship Mode']=0
df['total']=df['Consumer Total']+df['Corporate Total']+df['Home Office Total']
```

```python
df
```

```python
df.loc[df['Consumer Total'] >1, 'Segment'] = 'Consumer'
df.loc[df['Corporate Total'] >1, 'Segment'] = 'Corporate'
df.loc[df['Home Office Total'] >1, 'Segment'] = 'Home Office'
df.loc[df['Consumer First Class'] >1, 'Ship Mode'] = 'First Class'
df.loc[df['Consumer Same Day'] >1, 'Ship Mode'] = 'Same Day'
df.loc[df['Consumer Second Class'] >1, 'Ship Mode'] = 'Second Class'
df.loc[df['Consumer Standard Class'] >1, 'Ship Mode'] = 'Standard Class'
df.loc[df['Corporate First Class'] >1, 'Ship Mode'] = 'First Class'
df.loc[df['Corporate Same Day'] >1, 'Ship Mode'] = 'Same Day'
df.loc[df['Corporate Second Class'] >1, 'Ship Mode'] = 'Second Class'
df.loc[df['Corporate Standard Class'] >1, 'Ship Mode'] = 'Standard Class'
df.loc[df['Office First Class'] >1, 'Ship Mode'] = 'First Class'
df.loc[df['Office Same Day'] >1, 'Ship Mode'] = 'Same Day'
df.loc[df['Office Second Class'] >1, 'Ship Mode'] = 'Second Class'
df.loc[df['Office Standard Class'] >1, 'Ship Mode'] = 'Standard Class'
```
